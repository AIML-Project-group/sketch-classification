```python
from pathlib import Path
import shutil
```

```python
def _copy(self: Path, target):
    assert self.is_file()
    shutil.copy(self, target/(self.name + self.suffix))
Path.copy = _copy
```

```python
INPUT_DIR = '../input/cybertron-sketches/png'
INPUT = Path(INPUT_DIR)

OUTPUT = Path('./')

for folder in ('test', 'train'):
    if not (OUTPUT/folder).exists():
        (OUTPUT/folder).mkdir()
```

```python
for obj in INPUT.iterdir():
    if obj.is_file(): continue

    if not (OUTPUT/'train'/obj.name).exists():
        (OUTPUT/'train'/obj.name).mkdir()
        (OUTPUT/'test'/obj.name).mkdir()

    i = 0
    for img in (INPUT/obj.name).iterdir():
        if i < 64:
            img.copy(OUTPUT/'train'/obj.name)
        else:
            img.copy(OUTPUT/'test'/obj.name)
        i += 1
```
